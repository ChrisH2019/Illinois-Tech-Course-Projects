```python
import numpy as np

def compute_cos_dist(vec1, vec2, alpha, beta):
    """
    Compute the cosine distance between vec1 and vec2.
    """
    # Scale the vectors
    scale_vector = np.array([1, alpha, beta])
    vec1_scaled = vec1 * scale_vector
    vec2_scaled = vec2 * scale_vector

    # Compute cosine distance
    nu = np.dot(vec1_scaled, vec2_scaled)
    de = np.sqrt(np.dot(vec1_scaled, vec1_scaled)) * \
         np.sqrt(np.dot(vec2_scaled, vec2_scaled))
    cos = nu / de
    
    return cos
```

(a)

```python
def jaccard_dist(arr1, arr2, user1, user2):
    """
    Compute the Jaccard distance between two arrays.
    A larger Jaccard implies a larger distance.
    """
    n_intersection = np.dot(arr1, arr2)
    n_union = np.sum(np.where((arr1 + arr2) > 0, 1, 0))
    jaccard_similarity = n_intersection / n_union
    jaccard_distance = 1 - jaccard_similarity
    print('Jaccard distance between {} and {}: {}'.format(
          user1, user2, jaccard_distance))
```

```python
def cosine_dist(arr1, arr2, user1, user2):
    """
    Compute the cosine distance between two arrays.
    A larger (positive) cosine implies a smaller angle 
    and therefore a smaller distance.
    """
    # Compute cosine distance
    nu = np.dot(arr1, arr2)
    de = np.sqrt(np.dot(arr1, arr1)) * np.sqrt(np.dot(arr2, arr2))
    cos = nu / de
    print('The cosine distance between {} and {}: {}'.format(
          user1, user2, cos))
```

```python
# Extract user id, movie id and rating
user = data[['user_id', 'movie_id', 'rating']]

# Convert the ratings data into a utility matrix
utility = user.pivot_table(index='user_id', 
                        columns='movie_id', 
                        values='rating',
                        fill_value=0)

print('Number of users:', utility.shape[0])
print('Number of movies:', utility.shape[1])

utility.head()
```

```python
# Center user ratings
utility_centered = utility.copy(deep=True)
for user in range(1, utility.shape[0]+1):
    non_zero_ratings = (utility.loc[user] > 0).sum()
    if non_zero_ratings:
        tot_ratings = utility.loc[user].sum()
        mean_ratings = tot_ratings / non_zero_ratings
        # Only substract the nonbank entries from the mean
        for movie in range(1, utility.shape[1]+1):
            if utility.loc[user, movie] > 0:
                utility_centered.loc[user, movie] -= mean_ratings
```

```python
utility_centered.head()
```

```python
def cosine(vec1, vec2):
    """Compute the cosine similarity bewteen two vectors"""
    nu = np.dot(vec1, vec2)
    de = np.sqrt(np.dot(vec1, vec1)) * np.sqrt(np.dot(vec2, vec2))
    return nu / de

# Compute cosine similarity between user 1 and all others
sim = {}
for user in range(2, utility_centered.shape[0]+1):
    sim[user] = cosine(utility_centered.loc[1], utility_centered.loc[user])
```

```python
# Find the most similar users for user 1
top_10 = sorted(sim, key=sim.get, reverse=True)[:10]
print('The 10 most similar users for user 1:\n')
for user in top_10:
    print('User:', user, 'cosine similarity:', sim[user])
```

```python
# Calculate the expected rating for movie 508 for user 1
# based on the expected ratings from the 10 most similar user for user 1
tot_ratings_top_10 = 0
n_valid_top_10_ratings = 0
for user in top_10:
    # Check whether the top 10 user has rated the movie 508 or not
    if utility.loc[user, 508] > 0:
        n_valid_top_10_ratings += 1
        tot_ratings_top_10 += utility.loc[user, 508]
        
print('Number of ratings rated by the 10 most similar' + 
      ' users for movie 508 for user 1:', n_valid_top_10_ratings)
print('\nTotal valid ratings rated by the 10 most simila' + 
      ' users for movie 508 for user 1:', tot_ratings_top_10)
print('\nExpected rating for movie 508 for user 1:', 
      int(np.round(tot_ratings_top_10 / n_valid_top_10_ratings, 1)))
```
