```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
```

### Problem 1

```python
# Load iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

```python
import warnings
warnings.filterwarnings("ignore")

for i in range(1,6):
    # Create a classifier
    clf = DecisionTreeClassifier(min_samples_leaf=2, 
                                 min_samples_split=5, 
                                 max_depth=i, 
                                 random_state=0)

    # Train the model
    clf = clf.fit(X_train ,y_train)
    
    # Predict the test data
    y_pred = clf.predict(X_test)
    
    print('Max depth:', i)
    print(classification_report(y_test, y_pred))
    print('Confusion matrix:')
    print(confusion_matrix(y_test, y_pred))
    print()
    
    # Visualize data
    fig, ax = plt.subplots(figsize=(10, 10))
    plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True, filled=True)
    plt.show();
```

From depth **2 to 5**, the decision tree has the highest Recall score which is 0.93 in this context because it has the highest percentage of ground true labels are correctly predicted as true (a.k.a $\frac{tp}{tp+fn}$.

The decision tree with depth equals **1** has the lowest Precision score which is 0.50 in this context because it has the lowest percentage of predicted labels are ground true labels (a.k.a $\frac{tp}{tp+fp}$).

From depth **2 to 5**, the decision tree has the best F1 score which is 0.93 in this context.

Micro-average or micro for shor **collect decisions from all classes, compute contingence table and evalute the performance**; macro-average or macro for short **compute performance for each class, then average them**; weighted-average or weighted for shor **weight each class by the number of samples from that class**.

### Problem 2

```python
# Load the discrete version of Breast Cancer Wisconsin
bcw_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data'
col_names = ['sample_code_number',
             'clump_thickness',
             'uniformity_of_cell_size',
             'uniformity_of_cell_shape',
             'marginal_adhesion',
             'single_epithelial_cell_size',
             'bare_nuclei',
             'bland_chromatin',
             'normal_nucleoli',
             'mitoses',
             'class']
bcw = pd.read_csv(bcw_url, header=None, names=col_names)
bcw.head()
```

```python
bcw.info()
```

```python
# Transform bare_nuclie from object type to category
bcw['bare_nuclei'] = bcw['bare_nuclei'].astype('category')

# Transofrm class from int64 type to category
bcw['class'] = bcw['class'].astype('category')
```

```python
# Detect missing values
print('Before imputing:\n', pd.isna(bcw.replace('?', np.nan)).sum())
bcw.replace('?', np.nan, inplace=True)

# Impute missing values with mode
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
bcw.iloc[:,:] = imp.fit_transform(bcw)
print('\nAfter imputing:\n', pd.isna(bcw).sum())
```

```python
# Split the data into features and labels
X = bcw.iloc[:,:-1]
y = bcw.iloc[:,-1]
```

```python
# Induce a binary Decision Tree
clf = DecisionTreeClassifier(criterion='entropy',
                             min_samples_leaf=2, 
                             min_samples_split=5, 
                             max_depth=2, 
                             random_state=0)

# Train the model
clf = clf.fit(X, y)

# Visualize data
print('Splitting criterion: Entropy')
fig, ax = plt.subplots(figsize=(15, 15))
plot_tree(clf, feature_names=col_names, class_names=['benign', 'malignant'], rounded=True, filled=True, ax=ax)
plt.show();

y_pred = clf.predict(X)
print(classification_report(y, y_pred))
print('Confusion matrix:')
print(confusion_matrix(y, y_pred))
tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
print('FP:', fp, 'TP:', tp, 'FPR:', fp/(fp + tn), 'TPR:', tp/(tp+fn))
print()
```

**Statistics below are aquired from the above decision tree graph.**

```python
print('Root Entropy:', 0.929)
print('Left Child Entropy @ Level 2:', 0.184)
print('Right Child Entropy @ Level 2:', 0.614)
```

```python
info_gain = 0.929 - 429/699 * 0.184 - 270/699 * 0.614
print('Information Gain:', info_gain)
```

```python
print('Root Error Rate:', 241/699)
print('Left Child Error Rate @ Level 2:', 12/429)
print('Right Child Error Rate @ Level 2:', 41/270)
```

```python
error_rate = 241/699 - 429/699 * (12/429) - 270/699 * (41/270)
print('Error Rate Reduction:', error_rate)
```

**uniformity_of_cell_size** is selected for the first split and the decision boundary is **2.5**.
