### Step 1.2: Collecting Tweets on Keywords

```python
class MyStreamListener(tweepy.StreamListener):
    """
    This class is used to download tweets from the stream.
    """
    def __init__(self, api, num_tweets, file_name):
        self.api = api
        self.tweets_count = 0
        self.num_tweets = num_tweets
        self.file_name = file_name
        
    def on_status(self, status):
        tweet = status._json
        with open(self.file_name, 'a', encoding='utf-8') as file:
            file.write(json.dumps(tweet) + '\n')
        self.tweets_count += 1
        sys.stdout.write(
            "\rProgress: {}/{} tweets downloaded".format(
                self.tweets_count, self.num_tweets
            )
        )
        sys.stdout.flush()
        if self.tweets_count < self.num_tweets:
            return True
        else:
            return False

    def on_error(self, status):
        print(status.text)
```

### Step 1.4: Processing Tweets

Interested in tweets that are retweeted, quoted or both.

```python
# Make a copy of the original data set
tweets_df_copy = tweets_df.copy()
```

```python
def extract_tweets(df):
    """
    Extract 
        tweet datetime, 
        tweet user screen name, 
        tweet text or full tweet text if extended,
        retweeted user screen name if retweeted,
        retweeted text,
        quoted user screen name if quoted,
        quoted text.
    """
    tweets = []
    for idx, row in df.iterrows():
        tweet = {}
        tweet['created_at'] = pd.to_datetime(row['created_at'])
        tweet['tweet_user_screen_name'] = row['user']['screen_name']
        if not pd.isnull(row['extended_tweet']):
            tweet['tweet_text'] = row['extended_tweet']['full_text']
        else:
            tweet['tweet_text'] = row['text']
        if not pd.isnull(row['retweeted_status']):
            tweet['retweet_user_screen_name'] = row['retweeted_status']['user']['screen_name']
            tweet['retweet_text'] = row['retweeted_status']['text']
            tweet['retweet_count']=row['retweeted_status']['retweet_count']
        if not pd.isnull(row['quoted_status']):
            tweet['quote_user_screen_name'] = row['quoted_status']['user']['screen_name']
            tweet['quote_text'] = row['quoted_status']['text']
            tweet['quote_count']=row['quote_count']
        tweets.append(tweet)
    tweets_df = pd.DataFrame(tweets)
    return tweets_df
```

```python
tweets_processed = extract_tweets(tweets_df)
```

## Step 2: Data Visualization

### Step 2.1: Creating Retweet Network

```python
# Create directed retweet network
retweet_network = nx.from_pandas_edgelist(
    tweets_processed[tweets_processed.retweet_user_screen_name.notna()],
    source='retweet_user_screen_name',
    target='tweet_user_screen_name',
    create_using=nx.DiGraph)
```

```python
print('Node in retweet network:', len(retweet_network.nodes))
print('Edges in retweet network:', len(retweet_network.edges()))
```

### Step 2.2: Visualizing Retweet Network

```python
# Create size list
sizes = [x[1] for x in retweet_network.degree()]

# Draw the network
plt.figure(figsize=(12,12))
nx.draw_networkx(retweet_network, 
                 pos=nx.spring_layout(retweet_network), 
                 with_labels=False, 
                 node_size=sizes,
                 width=0.2, alpha=0.7,
                 arrowsize=3, linewidths=1)
plt.axis('off')
plt.title('Retweeted Network', fontsize=22)
plt.show()
```

```python
#visualizing the graph using the filtered data(sub dataframe)
G = nx.Graph()
G = nx.from_pandas_edgelist(final, 'retweet_user_screen_name', 'tweet_user_screen_name',create_using=nx.DiGraph())
from matplotlib.pyplot import figure
figure(figsize=(10, 5))
nx.draw_shell(G, with_labels=True)
```

```python
#visualizing the graph using filtered data(sub dataframe) with the normalized weighted edges
pos = nx.spring_layout(G)
plt.figure(figsize=(10,5))    
nx.draw(G,pos,edge_color='black',width=2,linewidths=3,\
node_size=100,node_color='blue',alpha=0.8,\
labels={node:node for node in G.nodes()})
nodes=[]
for node in G.nodes():
    nodes.append(node)
weight={}
for edge in G.edges():
    weight[edge[0],edge[1]]=list(final['retweet_count'].loc[(final['retweet_user_screen_name']==edge[0])&(final['tweet_user_screen_name']==edge[1])])[0]

nx.draw_networkx_edge_labels(G,pos,edge_labels=weight)
plt.axis('off')
plt.show()
```

```python
degrees = [val for (node, val) in retweet_network.degree()]

print(f"The maximum degree of the Graph is {np.max(degrees)}")   
print(f"The minimum degree of the Graph is {np.min(degrees)}")
```

## Step 3: Network Measures Calculation

### Step 3.1: In-degree Centrality for Retweet Network

```python
# Calculate in-degree centrality for retweet network
retweet_network_centrality = nx.in_degree_centrality(retweet_network)

# Store centralities into a dataframe
cols = ['screen_name', 'in_degree_centrality']
retweet_network_centrality = pd.DataFrame(
    list(retweet_network_centrality.items()), columns=cols)

#Print top 10 centralities in desending order
print('Top 10 Users with Highest In-Degree Centrality in Retweet Network\n')
retweet_network_centrality.sort_values('in_degree_centrality', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,8))
plt.hist(retweet_network_centrality.in_degree_centrality)
plt.title("Retweet Network In-Degree Centrality Distribution", fontsize=20)
plt.xlabel("In-Degree", fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.show()
```

### Step 3.2: Eigenvector Centrality for Retweet Network

```python
#calculate eigenvector centrality of the retweet network
centrality = nx.eigenvector_centrality(retweet_network)
values=[(node,round(centrality[node],4)) for node in centrality]
#display first 20 centrality values
values.sort(key=lambda x: x[1],reverse=True)
values[:20]
```

### Step 3.3: Katz Centrality for Retweet Network

```python
#calculate katz centrality of retweet network
centrality = nx.katz_centrality(retweet_network)
values=[(node,round(centrality[node],4)) for node in centrality]
#display first 20 centrality values
values.sort(key=lambda x: x[1],reverse=True)
values[:20]
```

### Step 3.4: Out-degree Centrality for Retweet Network

```python
# Calculate out-degree centrality for retweet network
retweet_network_centrality_out = nx.out_degree_centrality(retweet_network)

# Store centralities into a dataframe
cols = ['screen_name', 'out_degree_centrality']
retweet_network_centrality_out = pd.DataFrame(
    list(retweet_network_centrality_out.items()), columns=cols)

#Print top 10 centralities in desending order
print('Top 10 Users with Highest out-Degree Centrality in Retweet Network\n')
retweet_network_centrality_out.sort_values('out_degree_centrality', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,8))
plt.hist(retweet_network_centrality_out.out_degree_centrality)
plt.title("Retweet Network Out-Degree Centrality Distribution", fontsize=20)
plt.xlabel("Out-Degree", fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.show()
```

### Step 3.9: Keyword Frequency

```python
def count_keyword(word, data):
    """
    Check if a word is in a tweeted, retweeted or quoted tweet.
    """
    # fill null values with empty string
    df = data.fillna(value='')
    counts = df['tweet_text'].str.contains(word, case=False)
    counts |= df['retweet_text'].str.contains(word, case=False)
    counts |= df['quote_text'].str.contains(word, case=False)
    
    return counts
```

```python
tweets_processed['trump'] = count_keyword('trump', tweets_processed)
tweets_processed['biden'] = count_keyword('biden', tweets_processed)

# Print percentage of tweets mentioning 'Trump'
print("Percentage of tweets with #Trmup:", 
      np.round(tweets_processed['trump'].mean() * 100, 2))

# Print proportion of tweets mentioning 'Biden'
print("Percentage of tweets with #Biden:", 
      np.round(tweets_processed['biden'].mean() * 100, 2))
```
